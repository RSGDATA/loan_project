### The dataset was far too large to upload to GitHub. This notebook trims down unwanted data, then saves a random sample for use in later analysis.

```python
import pandas as pd
```

```python
accepted_df = pd.read_csv('accepted_2007_to_2018q4.csv/accepted_2007_to_2018q4.csv')
```

```python
accepted_df = accepted_df.drop(['id','member_id','funded_amnt','funded_amnt_inv','grade','sub_grade','emp_title',\
                               'issue_d','url','desc','purpose','earliest_cr_line','last_pymnt_d','next_pymnt_d',\
                               'last_credit_pull_d','mths_since_last_major_derog','policy_code'],axis=1)

```

```python
individual_df = accepted_df.loc[accepted_df['application_type']=='Individual']
individual_df.head(2)
```

```python
individual_df = individual_df.drop(['hardship_payoff_balance_amount','hardship_last_payment_amount','disbursement_method',\
                                    'debt_settlement_flag','debt_settlement_flag_date','settlement_status','settlement_date',\
                                    'settlement_amount','settlement_percentage','settlement_term'], axis=1)
individual_df.head(2)
```

```python
individual_df = individual_df.drop(['hardship_status','deferral_term','hardship_amount','hardship_start_date',\
                                    'hardship_end_date','payment_plan_start_date','hardship_length','hardship_dpd',\
                                    'hardship_loan_status','orig_projected_additional_accrued_interest'
], axis=1)

individual_df.head()
```

```python
individual_df = individual_df.drop(['sec_app_open_acc','sec_app_revol_util','sec_app_open_act_il','sec_app_num_rev_accts',\
                                    'sec_app_chargeoff_within_12_mths','sec_app_collections_12_mths_ex_med',\
                                    'sec_app_mths_since_last_major_derog','hardship_flag','hardship_type','hardship_reason'
], axis=1)

individual_df.head()
```

```python
individual_df = individual_df.drop(['revol_bal_joint','sec_app_fico_range_low','sec_app_fico_range_high',\
                                    'sec_app_earliest_cr_line','sec_app_inq_last_6mths','sec_app_mort_acc'
], axis=1)

individual_df.head()
```

```python
# Here, we're removing columns that have plenty of rows, but the titles are not thoroughly explained

individual_df = individual_df.drop(['application_type','verification_status_joint',\
                                    'open_acc_6m','open_act_il','open_il_12m','open_il_24m','mths_since_rcnt_il',\
                                    'total_bal_il','il_util','open_rv_12m','open_rv_24m','max_bal_bc','all_util'

], axis=1)

individual_df.head()
```

```python
individual_df['mths_since_recent_revol_delinq'] = individual_df['mths_since_recent_revol_delinq'].fillna(value=-1)
```

```python
individual_df['mths_since_recent_bc_dlq'] = individual_df['mths_since_recent_bc_dlq'].fillna(value=-1)
```

```python
c_list = ['inq_fi','total_cu_tl','inq_last_12m']

for c in c_list:
    individual_df[c] = individual_df[c].fillna(value=-1)
```

```python
accepted_df = individual_df
```

```python
columnslist = individual_df.columns
i = 0
for c in columnslist:
    print(f'{c:<43}', end='')
    i += 1
    if i % 3 == 0:
        print()
        
print(f'\n\nnumber of columns: {i}')
```

```python
accepted_sample = accepted_df.sample(frac=0.05, random_state=42)
```

```python
accepted_sample.head()
```

```python
accepted_sample.to_csv('accepted2.csv', index=False)
```
