```python
import pandas as pd
import numpy as np
```

```python
# Import csv to further clean data

rejected_df = pd.read_csv("../Resources/lending_club_rejected.csv")

rejected_df.head()
```

```python
# Zip code does not have full number and we have state to operate off of for charting

rejected_df = rejected_df.drop("Zip Code", axis = 1)
rejected_df.head()
```

```python
# Replacing string values with integers

rejected_df["Employment Length"] = rejected_df["Employment Length"].replace('< 1 year', 0.5).replace('10+ years', 10).\
replace('2 years', 2).replace('3 years', 3).replace('1 year', 1).replace('4 years', 4).replace('5 years', 5).\
replace('6 years', 6).replace('7 years', 7).replace('8 years', 8).replace('8 years', 8).replace('9 years', 9)

rejected_df.head()
```

```python
# drop NaN's from dataset
rejected_df = rejected_df.dropna()

rejected_df.head()
```

```python
rejected_df.reset_index(drop=True, inplace=True)
```

```python
# counting how many rows of data after cleaning (originally 691,219)

rejected_df.value_counts()
```

```python
rejected_df.head()
```

```python
# validating no NaN's remain

rejected_df.isnull().nunique()
```

```python
# correcting 'Debt-To-Income Ratio' from object to float

rejected_df['Debt-To-Income Ratio'] = rejected_df['Debt-To-Income Ratio'].str.replace('%','')
rejected_df = rejected_df.astype({
    'Debt-To-Income Ratio': float
})
```
